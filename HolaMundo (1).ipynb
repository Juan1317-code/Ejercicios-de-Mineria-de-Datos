```python
def factorial(n):
    resultado = 1

    for i in range(1, n + 1):
        resultado *= i

    return resultado

num = int(input("Ingresa un número: "))
print("Factorial:", factorial(num))
```

```python
def calcular_factura(cantidad, iva=21):
    total = cantidad + (cantidad * iva / 100)
    return total

cantidad = float(input("Ingresa la cantidad sin IVA: "))

print("Total con IVA:", calcular_factura(cantidad))
```

```python
import math

def area_circulo(radio):
    return math.pi * radio ** 2

def volumen_cilindro(radio, altura):
    return area_circulo(radio) * altura

radio = float(input("Radio: "))
altura = float(input("Altura: "))

print("Área del círculo:", area_circulo(radio))
print("Volumen del cilindro:", volumen_cilindro(radio, altura))
```
